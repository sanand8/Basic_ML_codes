```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import math
import random
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.utils import column_or_1d 
from sklearn import metrics
```

```python
derma_ds = pd.read_csv('C:/Users/welcome/Desktop/ML/Mod4/dermatology.data',header=None)
derma_ds.info()
```

```python
derma_ds[33] = pd.to_numeric(derma_ds[33], errors='coerce').fillna(0).astype(int)
derma_ds.info()
```

```python
x = derma_ds.iloc[:,0:34].values
y = derma_ds.iloc[:, 34].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
y_train = pd.DataFrame(y_train)
y_test = pd.DataFrame(y_test)
x_train[1:10]
#x_train[22]
#x_train.columns.tolist()
```

```python
#FORWARD ELIMINATION
significant_val = 0.05
initial_feature = x_train.columns.tolist()
best_feature = []
while(len(list(initial_feature)) > 0):
    rem_feature = list(set(initial_feature)-set(best_feature))
    new_pval = pd.Series(index=rem_feature)
    for new_col in rem_feature:
        #model = LogisticRegression()
        #model.fit(x_train[best_feature + [new_col]],y_train)
        #new_accuracy[new_col] = model.score(x_test[best_feature + [new_col]],y_test)
        model = sm.OLS(y_train, sm.add_constant(x_train[best_feature+[new_col]]))
        result = model.fit()
        new_pval[new_col] = result.pvalues[new_col]
    print(new_pval.min(),new_pval.idxmin())
    if(new_pval.min() < significant_val):
        best_feature.append(new_pval.idxmin())
    else:
        model = sm.OLS(y_train, sm.add_constant(x_train[best_feature]))
        result = model.fit()
        print('R2: ', result.rsquared)
        break
best_feature
```

```python
#BACKWARD ELIMINATION
feature = x_train.columns.tolist()
while(len(list(feature)) > 0):
    #features_const = sm.add_constant(x_train[feature])
    model = sm.OLS(y_train, sm.add_constant(x_train[feature]))
    result = model.fit().pvalues[1:]
    #new_pval[new_col] = result.pvalues[new_col]
    #print(new_pval.min(),new_pval.idxmin())
    if(result.max() >= significant_val):
        excluded_feature = result.idxmax()
        feature.remove(excluded_feature)
    else:
        print(result)
        break
feature
```
