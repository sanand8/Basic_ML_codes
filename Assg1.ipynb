```python
#importing the library's
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#reading data from file
data = pd.read_table('noise.dat')

X_0 = data.iloc[:, 0]
X_1 = data.iloc[:, 1]
X_2 = data.iloc[:, 2]
X_3 = data.iloc[:, 3]
X_4 = data.iloc[:, 4]

Y = data.iloc[:, 5]
plt.scatter(X_1,Y)
plt.show()
```

```python
#normalizing the data 
X_0 = (X_0 - X_0.mean()) / X_0.std()
X_1 = (X_1 - X_1.mean()) / X_1.std()
X_2 = (X_2 - X_2.mean()) / X_2.std()
X_3 = (X_3 - X_3.mean()) / X_3.std()
X_4 = (X_4 - X_4.mean()) / X_4.std()


#building the algorithm

m_0 = 0
m_1 = 0
m_2 = 0
m_3 = 0
m_4 = 0

c = 0
L = 0.01
iterq = 10099
J = np.ones(iterq)
n = float(len(X_0))

for i in range(iterq) :
    y_p = m_0 * X_0 + m_1* X_1+ m_2* X_2+ m_3* X_3+ m_4* X_4+ c
    
 
    #cost fn or error from the plotted points
    J[i] = (1/(2*n))*sum((y_p - Y)**2)
    
    #minimizing the slope and intercept
    D_m0= (1/n)* sum (X_0 * ( y_p - Y))
    D_m1= (1/n)* sum (X_1 * ( y_p - Y))
    D_m2= (1/n)* sum (X_2 * ( y_p - Y))
    D_m3= (1/n)* sum (X_3 * ( y_p - Y))
    D_m4= (1/n)* sum (X_4 * ( y_p - Y))
    D_c = (1/n)* sum (y_p - Y)
    
    m_0 = m_0 - L * D_m0
    m_1 = m_1 - L * D_m1
    m_2 = m_2 - L * D_m2
    m_3 = m_3 - L * D_m3
    m_4 = m_4 - L * D_m4
    
    c = c - L * D_c
    #print(J[i])
print(m_0,m_1,m_2,m_3,m_4, c)
#plt.scatter(iterq,J)
#plt.show()
plt.plot(range(iterq),J)
print(min(J))
print(J[998])
```
